```python
import numpy as np
import pandas as pd
import geopandas as gpd
from config import *
```

```python
# filepath
shapefile_filepath = INPUT_filepath+'03_shapefiles/'
```

```python
merged = []
for site in main_sites:
    # Load the shapefile
    shp_path = f"{shapefile_filepath}/{site}.shp"
    watershed = gpd.read_file(shp_path)
    watershed['STREAM_ID'] = site
    merged.append(watershed)

# Save as single shapefile
merged_gdf = gpd.GeoDataFrame(pd.concat(merged, ignore_index=True))
merged_gdf.drop(columns=['FID', '/web/group'], inplace = True)

# Reordering columns
col = merged_gdf.pop('STREAM_ID')
merged_gdf.insert(0, 'STREAM_ID', col)

merged_gdf.to_file(shapefile_filepath + 'merged_mrb_watersheds.shp')
```
