```python
import numpy as np
import pandas as pd
import geopandas as gpd
from config import *

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
nlcdClasses = {
    11: "Open Water",
    12: "Perennial Ice/Snow",
    21: "Developed Open Space",
    22: "Developed Low Intensity",
    23: "Developed Medium Intensity",
    24: "Developed High Intensity",
    31: "Barren Land",
    41: "Deciduous Forest",
    42: "Evergreen Forest",
    43: "Mixed Forest",
    52: "Shrub/Scrub",
    71: "Grassland/Herbaceous",
    81: "Pasture/Hay",
    82: "Cultivated Crops",
    90: "Woody Wetlands",
    95: "Emergent Herbaceous Wetlands"}
```

```python
# Averaging land use from 2008 (see config file) to end of avaialble LU data
MRB_lu_data = pd.read_csv(land_use_filepath + 'MRB_NLCD_LandCover_Pct_1985_2024.csv')
lu_data_save = []
for site in main_sites:
    site_lu_data = MRB_lu_data[MRB_lu_data['STREAM_ID'] == site]
    site_lu_data = site_lu_data[site_lu_data['year'] >= 2008].drop(columns=['year','STREAM_ID']).mean().to_frame().T
    site_lu_data.insert(0, 'STREAM_ID', site)
    lu_data_save.append(site_lu_data)

lu_data_export = pd.concat(lu_data_save, ignore_index=True)
lu_data_export.rename(columns={f'pct_{k}': v for k, v in nlcdClasses.items()}, inplace=True)
```

```python
lu_data_export.to_csv(land_use_filepath+'MRB_averaging_land_use.csv', index = False)
```
